# **What can we learn about different hosts and areas?**

> *Perform groupby function on host_name and areas and aggregation funtion on calculated_host_listings_count. we get the result, which tells us about the host_name and total number of listed properties at different neighbourhood_group.*


```python
# Performing groupby on host_name and calculated_hosts_listings and Sorting values by calculated_host_listings_count
hosts_listings = df.groupby(['host_name','neighbourhood_group'])['calculated_host_listings_count'].max().reset_index()
highest_calculated_host_listings_count = hosts_listings.sort_values(by='calculated_host_listings_count', ascending=False).head(10)
highest_calculated_host_listings_count
```

```python
# Using matplotlib libraries for ploting the bar chat of hosts with highest calculated host listings count
host_name = highest_calculated_host_listings_count['host_name']
calculated_host_listings_count = highest_calculated_host_listings_count['calculated_host_listings_count']
plt.bar(host_name,calculated_host_listings_count, color='red')

plt.rcParams['figure.figsize'] = (15,10)
plt.title('Hosts with Highest Host Listings', {'fontsize':15})
plt.xlabel('Host Names',{'fontsize':15})
plt.ylabel('Host listings count',{'fontsize':15})
plt.show()
```

In this following output we can see that the host_name *Sonder(NYC)* have most listing of 327 properties in *Manhattan Area*

```python
# Hosts and neighbourhood groups
hosts_per_neighbourhood_group = df.groupby(['neighbourhood_group'])['room_type'].count().reset_index().rename(columns={'room_type':'number_of_airbnb'}).sort_values(by='number_of_airbnb',ascending=False).head(10)
hosts_per_neighbourhood_group
```

```python
# Using matplotlib libraries for ploting the bar chart for easy visualize and understanding.
neighbourhood = hosts_per_neighbourhood_group['neighbourhood_group']
num_of_bnb = hosts_per_neighbourhood_group['number_of_airbnb']

fig = plt.figure(figsize = (15,5))
# Creating the bar plot
plt.bar(neighbourhood, num_of_bnb,color = ['darkgreen','green','limegreen','springgreen','lightgreen'])

plt.xlabel("Neighbourhood",fontsize=20)
plt.ylabel("number of Airbnb",fontsize=20)
plt.title('Number of hosts in each neighbourhood group',fontsize=20)
plt.xticks(fontsize='15')
plt.show()
```

**Observations:**

# **Which hosts are the busiest and why?**

```python
# Using groupby on host_id, host_name, room_type and neighbourhood_group with respect to maximun number of reviews
busiest_host = df.groupby(['host_name','room_type','neighbourhood_group'])['number_of_reviews'].max().reset_index().rename(columns={'number_of_reviews':'number_of_booking'})
busiest_host
```

```python
# Sort by number of bookings
top_10_busiest_host = busiest_host.sort_values(by='number_of_booking',ascending=False).head(10)
top_10_busiest_host
```

```python
# Using matplotlib libraries for ploting the bar chart for easy visualize and understanding to know how many bookings got the top hosts.
top_10_hosts_name = top_10_busiest_host['host_name']
num_booking = top_10_busiest_host['number_of_booking']

fig = plt.figure(figsize = (10, 5))

# Creating the bar plot
plt.bar(top_10_hosts_name,num_booking)
plt.xlabel("Host",fontsize=20)
plt.ylabel("Bookings",fontsize=20)
plt.title("Host vs Number of Bookings",fontsize=20)
plt.xticks(fontsize='12')
plt.show()
```

**Observations:**

*   Dona, Jj, Maya, Carol and Danielle are the busiest host.

# **Is there any noticeable difference of traffic among different areas and what could be the reason for it?**

> In order to know the traffic areas we have to mainly look into three columns.<br> Which are as follows:-
*   Neighbourhood group
*   Room type
*   Minimum nights






```python
# Getting the most traffic neighbourhood
trafic_areas = df.groupby('neighbourhood_group')['minimum_nights'].count().reset_index().sort_values(by='minimum_nights',ascending=False)
trafic_areas
```

```python
# Using matplotlib libraries for ploting the bar chart for easy visualize and understanding.
neighbourhood_type = trafic_areas['neighbourhood_group']
min_nights = trafic_areas['minimum_nights']

fig = plt.figure(figsize = (10, 5))
 
# Creating the bar plot
color = ['midnightblue','steelblue','cornflowerblue','mediumslateblue','lightsteelblue']
plt.bar(neighbourhood_type, min_nights,color=color)

plt.xlabel("Neighbourhood Group",fontsize=20)
plt.ylabel("Minimum number of nights stayed",fontsize=20)
plt.title("Traffic Areas in Different Neighbourhood",fontsize=20)
plt.xticks(fontsize='15')
plt.show()
```



---

##**Now checking what kind of room type people prefer in traffic area like manhattan?**

---



```python
# Groupby on neighbourhood_group and room_type with aggregation on ninimun_nights 
traffic_among_diff_areas = df.groupby(['neighbourhood_group','room_type'])['minimum_nights'].count().reset_index()
traffic_among_diff_areas = traffic_among_diff_areas.sort_values(by='minimum_nights', ascending=False)
traffic_among_diff_areas
```

```python
# Getting only room_type and minimun_nights data
room_type_vs_night = traffic_among_diff_areas[traffic_among_diff_areas['neighbourhood_group']=='Manhattan'].groupby('room_type')['minimum_nights'].sum().reset_index()
room_type_vs_night
```

```python
# Using matplotlib libraries for ploting the bar chart for easy visualize and understanding.
room_type = room_type_vs_night['room_type']
stayed = room_type_vs_night['minimum_nights']

fig = plt.figure(figsize = (10, 5))
 
# Creating the bar plot
color = ['blue','royalblue','cornflowerblue']
plt.bar(room_type, stayed,color=color)


plt.xlabel("Room Type",fontsize=20)
plt.ylabel("Minimum number of nights stayed",fontsize=20)
plt.title("Traffic Areas",fontsize=20)
plt.xticks(fontsize='15')
plt.show()
```

```python
room_type_vs_night_in_manhattan = traffic_among_diff_areas[traffic_among_diff_areas['neighbourhood_group']=='Manhattan'].groupby('room_type')['minimum_nights'].sum().reset_index()
room_type_vs_night_in_manhattan
```

```python
room_type_vs_night_in_brooklyn = traffic_among_diff_areas[traffic_among_diff_areas['neighbourhood_group']=='Brooklyn'].groupby('room_type')['minimum_nights'].sum().reset_index()
room_type_vs_night_in_brooklyn
```

```python
room_type_vs_night_in_queens = traffic_among_diff_areas[traffic_among_diff_areas['neighbourhood_group']=='Queens'].groupby('room_type')['minimum_nights'].sum().reset_index()
room_type_vs_night_in_queens
```

```python
room_type_vs_night_in_bronx = traffic_among_diff_areas[traffic_among_diff_areas['neighbourhood_group']=='Bronx'].groupby('room_type')['minimum_nights'].sum().reset_index()
room_type_vs_night_in_bronx
```

```python
room_type_vs_night_in_statenisland = traffic_among_diff_areas[traffic_among_diff_areas['neighbourhood_group']=='Staten Island'].groupby('room_type')['minimum_nights'].sum().reset_index()
room_type_vs_night_in_statenisland
```

**Observations:**

*   'Entire home/apt' type has more demanding in traffic areas as compare to 'private room' and 'shared room' and 'Entire home/apt' usually booked by family and who has travel for short business trip along with family.
*   'Private room' type usually booked by bachelor's, Couples or Travellers who don't wanted to share privacy with others.
*   'Shared room' type mostly booked by students who does't earn much and don't wanted to spend much price for stay.
